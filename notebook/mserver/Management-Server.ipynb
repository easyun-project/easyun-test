```python
# 查看Region内所有云服务器
import boto3

REGION = "us-east-1"
FLAG = 'Easyun'
RESOURCE = boto3.resource('ec2', region_name=REGION)

servers = RESOURCE.instances.all()
for server in servers:
    print(f'ID: {server.id}')
    print(f'Server type: "{server.instance_type}')
    print(f'Server AMI: {server.image.id}')
#    print(f'Server platform: {server.platform}')
    print(f'Piblic(IPv4): {server.public_ip_address}')
    print(f'Server state: {server.state["Name"]}')    
    print('-'*50)
```

```python
# 查看指定服务器的完整详细信息

import json
from datetime import date, datetime
import boto3

REGION = "us-east-1"
FLAG = 'Easyun'
CLIENT = boto3.client('ec2', region_name=REGION)
ID = 'i-0a3ee56a27d528525'
# ID = 'i-043e8074448089cad'


# Helper method to serialize datetime fields
def json_datetime_serializer(obj):
    if isinstance(obj, (datetime, date)):
        return obj.isoformat()
    raise TypeError ("Type %s not serializable" % type(obj))

response = CLIENT.describe_instances(
    InstanceIds=[ID],
)

print(f'Server attributes ({ID}):')

for reservation in response['Reservations']:
    print(json.dumps(
            reservation,
            indent=4,
            default=json_datetime_serializer
        )
    )
```

```python
# 修改指定服务器的配置/属性
import boto3

REGION = "us-east-1"
FLAG = 'Easyun'
RESOURCE = boto3.resource('ec2', region_name=REGION)
ID = 'i-010d56cdd684dd27a'

server = RESOURCE.Instance(ID)

#修改实例类型需停止服务器
server.stop()
server.wait_until_stopped()
print(f'Status: {server.id} has been stoped')

server.modify_attribute(
    InstanceType={
        'Value': 't4g.nano'
    }
)

#修改完成后启动服务器
server.start()
server.wait_until_running()

print(f'Status: Instance type has been successfully changed')
```

```python
# 服务器基本操作 - 删除
import boto3

RESOURCE = boto3.resource('ec2')

IDs = ['i-010d56cdd684dd27a', 'i-05601c9a6d3dbe7c4']

# server = RESOURCE.Instance(ID)
# ec2.instances.filter(InstanceIds=ids).stop()

# Terminating instances
servers = RESOURCE.instances.filter(InstanceIds=IDs)

for server in servers:
    server.terminate()
    print(f'Terminating Server: {server.id}')  

    server.wait_until_terminated()
    print(f'Server "{server.id}" has been terminated')
```

```python
# [Option]添加/修改tags
import boto3

REGION = "us-east-1"
FLAG = 'Easyun'
RESOURCE = boto3.resource('ec2', region_name=REGION)

IDs = ['i-0a3ee56a27d528525', ]
TAGS = [
    {
        'Key': 'Env', 
        'Value': 'Test'
    },
    {
        'Key': 'Flag', 
        'Value': 'EasyunTest'
    }
]

servers = RESOURCE.instances.filter(InstanceIds=IDs)
for server in servers:
    server.create_tags(Tags=TAGS)
    for tag in server.tags:
        print(f'- Tag: {tag["Key"]}={tag["Value"]}')
```
